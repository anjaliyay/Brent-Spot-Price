```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Given dataset
data = {
    "Year-Month": ["1990-01", "1990-02", "1990-03", "1990-04", "1990-05", "1990-06", "1990-07", "1990-08", "1990-09", "1990-10", "1990-11"],
    "Price (USD/barrel)": [21.251, 19.813, 18.387, 16.612, 16.352, 15.104, 17.169, 27.167, 34.899, 36.022, 33.067]
}
```

```python
# Convert to DataFrame
df = pd.DataFrame(data)
```

```python
# Convert 'Year-Month' column to datetime format and set it as index
df['Year-Month'] = pd.to_datetime(df['Year-Month'])
df.set_index('Year-Month', inplace=True)
```

```python
# Plot the trend
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Price (USD/barrel)'], marker='o', color='b', linestyle='-')
plt.title('Trend Analysis of Brent Crude Oil Spot Price')
plt.xlabel('Year-Month')
plt.ylabel('Price (USD/barrel)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
